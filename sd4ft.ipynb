```python
import pandas as pd
from cleverminer import cleverminer
```

```python
data = pd.read_csv('crime_stations_transformed2.csv', sep=',')

df = pd.DataFrame(data)

df['Daytime_num'] = df['Daytime'].apply(lambda x: 1 if x == 'Early Morning' else (2 if x == 'Morning' else (3 if x=='Afternoon' else(4 if x=='Evening' else 5))))
df.head()
```

```python
clm = cleverminer(df=df,proc='SD4ftMiner',
               quantifiers= {'Base1':100, 'Base2':100, 'Ratioconf' : 1.5},
               ante ={
                    'attributes':[
                        {'name': 'Description', 'type': 'subset', 'minlen': 1, 'maxlen': 1},
                        {'name': 'Weapon', 'type': 'subset', 'minlen': 1, 'maxlen': 1},
                        {'name': 'Gender', 'type': 'subset', 'minlen': 1, 'maxlen': 1},
                        {'name': 'Age_Category', 'type': 'subset', 'minlen': 1, 'maxlen':1},
                        {'name': 'Police_in_neigh', 'type': 'subset', 'minlen': 1, 'maxlen': 1}
                    ], 'minlen':1, 'maxlen':5, 'type':'con'},
               succ ={
                    'attributes':[
                        {'name': 'Daytime_num', 'type': 'rcut', 'minlen': 1, 'maxlen': 2}
                    ], 'minlen':1, 'maxlen':1, 'type':'con'},
               frst ={
                    'attributes':[
                        {'name': 'Old_District', 'type': 'subset', 'minlen': 1, 'maxlen': 1}
                    ], 'minlen':1, 'maxlen':1, 'type':'con'},
               scnd ={
                    'attributes':[
                        {'name': 'Old_District', 'type': 'subset', 'minlen': 1, 'maxlen': 1}
                    ], 'minlen':1, 'maxlen':1, 'type':'con'}
               )


clm.print_summary()
clm.print_rulelist()
clm.print_rule(1)
```
