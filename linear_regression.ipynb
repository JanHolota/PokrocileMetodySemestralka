```python
# nahrání potřebných knihoven
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_squared_error
```

```python
data = pd.read_csv('crime_stations_transformed2.csv', sep=',')

df= pd.DataFrame(data)
df.head()
```

```python
# uložíme si vysvětlující i vysvětlovanou proměnnou
x = df['Hour']
y = df['distance_to_police']
```

```python
# rozdělíme dataset na trénovací a testovací část
train_x,test_x,train_y,test_y = train_test_split(x,y,test_size=0.2)
```

```python
# vytvoříme regresní model
model = LinearRegression()

# přetvoříme dataframe s vysvětlující proměnnou na 2D pole, implementace modelu to vyžaduje
train_x_2d = train_x.values.reshape(-1, 1)
test_x_2d = test_x.values.reshape(-1, 1)

# natrénování modelu
model.fit(train_x_2d, train_y)

# vytvoření predikční pro testovací sadu
prediction = model.predict(test_x_2d)
```

```python
# dle r2 i MSE je jasné, že lineární vztah mezi těmito proměnnými není.
r2 = r2_score(test_y,prediction)
mse = mean_squared_error(test_y,prediction)
print(r2,mse)
```
