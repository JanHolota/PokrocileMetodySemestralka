```python
import pandas as pd
from cleverminer import cleverminer
```

```python
data = pd.read_csv('crime_stations_transformed2.csv', sep=',')

df = pd.DataFrame(data)

## aby šel udělat right cut na datech, je potřeba jednotlivé kategorie nahradit číselnými hodnotami
df['Age_Category_num'] = df['Age_Category'].apply(lambda x: 1 if x == '0-18' else (2 if x == '19-25' else (3 if x=='26-35' else (4 if x=='36-45' else(5 if x=='46-55' else (6 if x=='56-65' else (7 if x=='66-75' else 8)))))))

df.head()
```

```python
clm = cleverminer(df=df,proc='SD4ftMiner',
               quantifiers= {'Base1':1000, 'Base2':1000, 'Ratioconf' : 1.6},
               ante ={
                    'attributes':[
                        {'name': 'Description', 'type': 'subset', 'minlen': 1, 'maxlen': 1},
                        {'name': 'Weapon', 'type': 'subset', 'minlen': 1, 'maxlen': 1},
                        {'name': 'Gender', 'type': 'subset', 'minlen': 1, 'maxlen': 1},
                        {'name': 'Daytime', 'type': 'subset', 'minlen': 1, 'maxlen':1},
                        {'name': 'Police_in_neigh', 'type': 'subset', 'minlen': 1, 'maxlen': 1}
                    ], 'minlen':1, 'maxlen':5, 'type':'con'},
               succ ={
                    'attributes':[
                        {'name': 'Age_Category_num', 'type': 'rcut', 'minlen': 1, 'maxlen': 2}
                    ], 'minlen':1, 'maxlen':1, 'type':'con'},
               frst ={
                    'attributes':[
                        {'name': 'Old_District', 'type': 'subset', 'minlen': 1, 'maxlen': 1}
                    ], 'minlen':1, 'maxlen':1, 'type':'con'},
               scnd ={
                   'attributes':[
                        {'name': 'Old_District', 'type': 'subset', 'minlen': 1, 'maxlen': 1}
                    ], 'minlen':1, 'maxlen':1, 'type':'con'}
                    
               )


clm.print_summary()
clm.print_rulelist()
clm.print_rule(1)
```
