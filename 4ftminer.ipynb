```python
import pandas as pd
import sys

from cleverminer import cleverminer

df = pd.read_csv('crime_stations_transformed2.csv', encoding='utf-8', sep=',')
df = df[['Description', 'Weapon', 'PremiseType', 'Daytime', 'Month', 'Police_in_neigh']]


clm = cleverminer(df=df, proc='4ftMiner', quantifiers={'conf': 0.9, 'Base': 10},
                  ante={
                    'attributes': [
                        {'name': 'Description', 'type': 'subset', 'minlen': 1, 'maxlen': 1},
                        {'name': 'Weapon', 'type': 'subset', 'minlen': 1, 'maxlen': 1},
                        {'name': 'PremiseType', 'type': 'subset', 'minlen': 1, 'maxlen': 1},
                        {'name': 'Daytime', 'type': 'subset', 'minlen': 1, 'maxlen': 1},
                        {'name': 'Month', 'type': 'seq', 'minlen': 1, 'maxlen': 3}
                    ], 'minlen': 5, 'maxlen': 5, 'type': 'con'
                  },
                  succ={
                    'attributes': [
                        {'name': 'Police_in_neigh', 'type': 'one', 'value': 1}
                    ], 'minlen': 1, 'maxlen': 1, 'type': 'con'
                  },
                  )

clm.print_summary()
clm.print_rulelist()
clm.print_rule(1)
```

```python
import pandas as pd
import sys

from cleverminer import cleverminer

df = pd.read_csv('crime_stations_transformed2.csv', encoding='utf-8', sep=',')
df = df[['Age_Category', 'Gender', 'Race', 'Daytime', 'Description', 'Police_in_neigh']]


clm = cleverminer(df=df, proc='4ftMiner', quantifiers={'conf': 0.7, 'Base': 50},
                  ante={
                    'attributes': [
                        {'name': 'Race', 'type': 'subset', 'minlen': 1, 'maxlen': 1},
                        {'name': 'Gender', 'type': 'subset', 'minlen': 1, 'maxlen': 1},
                        {'name': 'Age_Category', 'type': 'seq', 'minlen': 1, 'maxlen': 2}
                    ], 'minlen': 3, 'maxlen': 3, 'type': 'con'
                  },
                  succ={
                    'attributes': [
                        {'name': 'Daytime', 'type': 'subset', 'minlen': 1, 'maxlen': 2}
                    ], 'minlen': 1, 'maxlen': 1, 'type': 'con'
                  },
                  cond={
                    'attributes': [
                        {'name': 'Police_in_neigh', 'type': 'one', 'value': 1},
                        {'name': 'Description', 'type': 'one', 'value': 'LARCENY'}
                    ], 'minlen': 2, 'maxlen': 2, 'type': 'con'
                  }
                  )

clm.print_summary()
clm.print_rulelist()
clm.print_rule(1)
```
