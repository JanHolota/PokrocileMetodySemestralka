```python
import pandas as pd
import sys

from cleverminer import cleverminer

df = pd.read_csv ('crime_stations_transformed.csv', encoding='cp1250', sep=',', header = 0)


clm = cleverminer(df=df,target='Daytime',proc='CFMiner',
               quantifiers= {'RelMax':0.25},
               cond ={
                    'attributes':[
                        {'name': 'Old_District', 'type': 'subset', 'minlen': 1, 'maxlen': 1},
                    ], 'minlen':1, 'maxlen':1, 'type':'con'},
                opts = {'max_categories': 1000}
               )


clm.print_summary()
clm.print_rulelist()
clm.print_rule(1)
```

```python
import pandas as pd
import sys

from cleverminer import cleverminer

df = pd.read_csv ('crime_stations_transformed.csv', encoding='cp1250', sep=',', header = 0)


clm = cleverminer(df=df, target='Month', proc='CFMiner',
               quantifiers= {'base':30000,'RelMin':0.05, 'RelMax_leq':0.10},
               cond ={
                    'attributes':[
                        {'name': 'Gender', 'type': 'subset', 'minlen': 1, 'maxlen': 1},
                        {'name': 'Age_Category', 'type': 'subset', 'minlen': 1, 'maxlen': 1},
                        {'name': 'Race', 'type': 'subset', 'minlen': 1, 'maxlen': 1},
                    ], 'minlen':3, 'maxlen':3, 'type':'con'},
               )


clm.print_summary()
clm.print_rulelist()
clm.print_rule(1)

```
